```python
import math
import numpy as np
os.makedirs('./output1')
os.makedirs('./output1/train')
os.makedirs('./output1/test')

root_dir = './output'
classes = os.listdir('output')

for clss in classes:
  dirtry = root_dir + '/' + clss
  files = os.listdir(dirtry)
  np.random.shuffle(files)

  base_outdir = './output1/'

  for folder in ['train', 'test']:
      target_dir = base_outdir + folder
      os.makedirs(target_dir + '/' + clss)
      target_class = target_dir + '/' + clss

      if folder == 'train':
        images_to_pass = files[: math.floor(0.75*len(files))]
        for img in images_to_pass:
            img = dirtry + '/' + img
            shutil.copy(img, target_class)
      else:
        images_to_pass = files[math.floor(0.75*len(files)):]
        for img in images_to_pass:
            img = dirtry + '/' + img
            shutil.copy(img, target_class)
```

```python
import keras

img_size = (128,128)
batch_size = 20
```

```python
vgg = keras.applications.vgg19.VGG19(weights='imagenet',include_top=False)
vgg.summary()
```

```python
#data_gen = keras.preprocessing.image.ImageDataGenerator()
data_gen = keras.preprocessing.image.ImageDataGenerator(rescale=1./255.)
train = data_gen.flow_from_directory('./output1/train', target_size=img_size, shuffle=False, batch_size=20)
test = data_gen.flow_from_directory('./output1/test', target_size=img_size, shuffle=False, batch_size=20)
```

```python
train_features = vgg.predict_generator(train,steps=5542//20)
test_features = vgg.predict_generator(test,steps=1848//20)
```

```python
print(len(train_features))
print(len(test_features))
```

```python
tr = os.listdir('./output1/train')
t = os.listdir('./output1/test')
l_tr = []
l_t = []
for dir in tr:
  l_tr.append(len(os.listdir(os.path.join('./output1/train', dir))))
for dir1 in t:
  l_t.append(len(os.listdir(os.path.join('./output1/test', dir1))))
l_tr = np.array(l_tr)
l_t = np.array(l_t)
```

```python
train_labels = np.array([0]*l_tr[0]+[1]*l_tr[1]+[2]*l_tr[2]+[3]*l_tr[3]+[4]*l_tr[4]+[5]*l_tr[5]+[6]*l_tr[6]+[7]*l_tr[7]+\
[8]*l_tr[8]+[9]*l_tr[9]+[10]*l_tr[10]+[11]*l_tr[11]+[12]*l_tr[12]+[13]*l_tr[13]+[14]*l_tr[14]+[15]*l_tr[15]+\
[16]*l_tr[16]+[17]*l_tr[17]+[18]*l_tr[18]+[19]*l_tr[19]+[20]*l_tr[20]+[21]*l_tr[21]+[22]*l_tr[22]+[23]*l_tr[23]+\
[24]*l_tr[24]+[25]*l_tr[25]+[26]*l_tr[26]+[27]*l_tr[27]+[28]*l_tr[28]+[29]*l_tr[29]+[30]*l_tr[30]+[31]*l_tr[31]+\
[32]*l_tr[32]+[33]*l_tr[33]+[34]*l_tr[34]+[35]*l_tr[35]+[36]*(l_tr[36]-4))
test_labels = np.array([0]*l_t[0]+[1]*l_t[1]+[2]*l_t[2]+[3]*l_t[3]+[4]*l_t[4]+[5]*l_t[5]+[6]*l_t[6]+[7]*l_t[7]+\
[8]*l_t[8]+[9]*l_t[9]+[10]*l_t[10]+[11]*l_t[11]+[12]*l_t[12]+[13]*l_t[13]+[14]*l_t[14]+[15]*l_t[15]+\
[16]*l_t[16]+[17]*l_t[17]+[18]*l_t[18]+[19]*l_t[19]+[20]*l_t[20]+[21]*l_t[21]+[22]*l_t[22]+[23]*l_t[23]+\
[24]*l_t[24]+[25]*l_t[25]+[26]*l_t[26]+[27]*l_t[27]+[28]*l_t[28]+[29]*l_t[29]+[30]*l_t[30]+[31]*l_t[31]+\
[32]*l_t[32]+[33]*l_t[33]+[34]*l_t[34]+[35]*l_t[35]+[36]*(l_t[36]-9))
print(len(train_labels))
print(len(test_labels))
```

```python
from keras.layers import *
model = keras.models.Sequential()
model.add(Flatten(input_shape=train_features.shape[1:]))
model.add(Dense(512, activation='softsign'))
model.add(Dropout(0.5))
model.add(Dense(37, activation='softmax'))

model.compile(optimizer='rmsprop',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(train_features, train_labels,
          epochs=25,
          batch_size=20,
          validation_data=(test_features, test_labels))
```

```python
import functools
top_k_acc = functools.partial(keras.metrics.top_k_categorical_accuracy, k=3)

top_k_acc.__name__ = 'top_k_acc'

model.compile(optimizer='rmsprop',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy','top_k_categorical_accuracy',top_k_acc])

model.evaluate(test_features, test_labels)
```

```python
top_k_acc = functools.partial(keras.metrics.top_k_categorical_accuracy, k=5)

top_k_acc.__name__ = 'top_k_acc'

model.compile(optimizer='rmsprop',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy','top_k_categorical_accuracy',top_k_acc])

model.evaluate(test_features, test_labels)
```

```python
predictions = model.predict(test_features).argmax(axis=1)

cnf = confusion_matrix(test_labels, predictions)

for i in range(len(cnf)):
  print(cnf[i][:])
  print()
```

```python
data_gen = keras.preprocessing.image.ImageDataGenerator(rescale=1./255.)
train = data_gen.flow_from_directory('./output3/train', target_size=img_size, shuffle=False, batch_size=20, class_mode='binary')
test = data_gen.flow_from_directory('./output3/test', target_size=img_size, shuffle=False, batch_size=20, class_mode='binary')
```

```python
train_features2 = vgg.predict_generator(train,steps=5542//20)
test_features2 = vgg.predict_generator(test,steps=1848//20)
```

```python
print(len(train_features2))
print(len(test_features2))
```

```python
tr = os.listdir('./output3/train')
t = os.listdir('./output3/test')
l_tr = []
l_t = []
for dir in tr:
  l_tr.append(len(os.listdir(os.path.join('./output3/train', dir))))
for dir1 in t:
  l_t.append(len(os.listdir(os.path.join('./output3/test', dir1))))
l_tr = np.array(l_tr)
l_t = np.array(l_t)
```

```python
train_labels2 = np.array([0]*l_tr[0]+[1]*(l_tr[1]-4))
test_labels2 = np.array([0]*l_t[0]+[1]*(l_t[1]-9))
print(len(train_labels2))
print(len(test_labels2))
```

```python
model = keras.models.Sequential()
model.add(Flatten(input_shape=train_features2.shape[1:]))
model.add(Dense(512, activation='softsign'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(train_features2, train_labels2,
          epochs=15,
          batch_size=20,
          validation_data=(test_features2, test_labels2))
```

```python
predictions = model.predict(test_features2).argmax(axis=1)

cnf = confusion_matrix(test_labels2, predictions)

cnf
```
